## What is a Conditional Statement?

In real life, we make decisions based on conditions.

Example: **If it is raining, carry an umbrella. Otherwise, enjoy the weather.**

Python works the same way. A conditional statement allows a program to decide which block of code should run based on whether a condition is `True` or `False`.


## Basic Syntax

```python
if condition:
    # code runs when condition is True
else:
    # code runs when condition is False
```

Indentation is very important in Python. The code inside an `if`, `elif`, or `else` block must be indented, usually by 4 spaces.


## 1. `if` Statement

The `if` statement is the simplest conditional statement. The code block runs only when the condition is `True`.


````python
age = 25

if age >= 18:
    print("You are eligible to vote.")
````

## 2. `if-else` Statement

Use `else` when you want one block to run if the condition is `True`, and another block to run if the condition is `False`.


````python
age = 16

if age >= 18:
    print("You can vote.")
else:
    print("You cannot vote.")
````

## 3. `if-elif-else` Statement

Use `elif` when you have more than two possible outcomes. Python checks conditions from top to bottom and runs the first block whose condition is `True`.


````python
marks = 82

if 90 <= marks <= 100:
    print("Grade: A")
elif 75 <= marks < 90:
    print("Grade: B")
elif 60 <= marks < 75:
    print("Grade: C")
elif 40 <= marks < 60:
    print("Grade: D")
elif 0 <= marks < 40:
    print("Grade: F")
else:
    print("Invalid marks")
````

## 4. Nested `if-else`

Nested conditions mean writing one `if` statement inside another. This is useful when a second condition should only be checked after the first condition is already true.


````python
age = 22
has_degree = True

if age >= 18:
    print("Age requirement met.")
    if has_degree:
        print("You are eligible for this job.")
    else:
        print("You need a degree for this job.")
else:
    print("You are too young to apply.")
````

## 5. Ternary Operator

The ternary operator lets you write a simple `if-else` condition in one line.


````python
age = 19

# Normal way
if age >= 18:
    status = "Adult"
else:
    status = "Minor"

print(status)

# Ternary operator
status = "Adult" if age >= 18 else "Minor"
print(status)
````

## Question 23

````python
price = float(input("Enter product price: "))
discount = float(input("Enter discount percentage: "))

if price < 0:
    print("Invalid price")
elif 1 <= discount <= 50:
    final_price = price - (price * discount / 100)
    print("Final price:", final_price)
else:
    print("Invalid discount. No discount applied.")
    print("Final price:", price)
````

## Question 26

Accept two numbers and an operator. Perform calculation using `if-elif-else` and handle division by zero.


````python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter an operator (+, -, *, /): ").strip()

if operator == "+":
    print(num1 + num2)
elif operator == "-":
    print(num1 - num2)
elif operator == "*":
    print(num1 * num2)
elif operator == "/":
    if num2 == 0:
        print("Error: Division by zero")
    else:
        print(num1 / num2)
else:
    print("Invalid operator")
````
