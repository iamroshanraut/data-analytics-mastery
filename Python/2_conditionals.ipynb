## 1. What is a Conditional Statement?

In real life we make decisions based on conditions:  
*"If it is raining, carry an umbrella — otherwise don't."*

Python works the same way. A **conditional statement** lets your program make decisions and execute different code depending on whether a condition evaluates to `True` or `False`.

| Keyword | Purpose |
|---------|----------------------------------------------------|
| `if` | Run a block only when the condition is `True` |
| `elif` | Check another condition if the previous one failed |
| `else` | Run a block when all conditions above are `False` |

````python
# Basic example – deciding based on a variable value
age = 25

if age >= 18:
    print("You are eligible to vote.")
````

## 2. `if` Statement

The simplest form. If the condition is `True`, the indented block runs.  
If it is `False`, Python **skips** it entirely.

```
if <condition>:
    <code to run when True>
```

> ⚠️ **Indentation is mandatory in Python.** The block inside `if` must be indented (usually 4 spaces). Missing indentation raises an `IndentationError`.

````python
marks = 75

if marks >= 40:
    print("You have passed.")

# If marks were 30, nothing would print – the block is simply skipped.
````

## 3. `if-else` Statement

Use `else` when you want *something* to happen when the condition is `False`.  
**Exactly one** of the two blocks always runs — never both, never neither.

```
if <condition>:
    <runs when True>
else:
    <runs when False>
```

````python
age = 16

if age >= 18:
    print("You can vote.")
else:
    print("You can't vote yet.")
````

## 4. `if-elif-else` Statement

When you have **more than two** possible outcomes, use `elif` (short for *else if*).  
Python checks each condition **top to bottom** and runs the first one that is `True`. The rest are skipped.

```
if <condition1>:
    ...
elif <condition2>:
    ...
elif <condition3>:
    ...
else:
    ...
```

````python
marks = 82

if marks >= 91:
    print('Grade: A')
elif marks >= 75:
    print('Grade: B')   # ← this branch runs for marks = 82
elif marks >= 61:
    print('Grade: C')
elif marks >= 41:
    print('Grade: D')
elif marks >= 0:
    print('Grade: F')
else:
    print('Invalid marks')

# Tip: once Python finds a True branch, all remaining elif/else are skipped.
# Writing marks >= 91 first (most restrictive) keeps the logic clean.
````

## 5. Nested Conditionals

You can place an `if` statement **inside** another `if` statement.  
This is called **nesting** and is useful when a second condition only makes sense if the first one is already `True`.

> 💡 Avoid nesting more than 2–3 levels deep — it hurts readability. Prefer logical operators (`and`, `or`) when possible.

````python
age = 22
has_degree = True

if age >= 18:
    print('Age requirement met.')
    if has_degree:
        print('You are eligible for this job.')
    else:
        print('You need a degree for this job.')
else:
    print('You are too young to apply.')
````

## 6. Ternary Operator (Shorthand `if-else`)

Python lets you write a simple `if-else` on a **single line**.  
This is called the **ternary operator** (or conditional expression).

```python
value = <true_result> if <condition> else <false_result>
```

> ✅ Use it for **simple, readable** one-liners.  
> ❌ Avoid chaining multiple ternary operators — it becomes unreadable.

````python
age = 19

# ── Standard if-else ──────────────────────────────────────────
if age >= 18:
    status = "adult"
else:
    status = "minor"
print(status)

# ── Ternary equivalent (same result, one line) ─────────────────
status = "adult" if age >= 18 else "minor"
print(status)
````

### Q43 · Short-Circuit Evaluation
Explain short-circuit evaluation. Show a case where the second condition is **never evaluated**.

````python
# In Python, `and` short-circuits: if the first operand is False,
# the second operand is NEVER evaluated.

age     = 17
pass_id = True

# Because age >= 18 is False, Python never checks pass_id.
if age >= 18 and pass_id:
    print('Allowed')
else:
    print('Not allowed (age check failed — pass_id was never evaluated)')

# Similarly, `or` short-circuits: if the first is True, the second is skipped.
x = 10
if x > 5 or (1 / 0):  # ZeroDivisionError never raised!
    print('Short-circuit saved us from a ZeroDivisionError.')
````

### Q44 · Operator Precedence: `or` vs `and`
`and` has **higher precedence** than `or`. Analyse the output.

````python
x = 5
y = 10

# Expression: x == 5 or y == 5 and y == 10
# Parsed as:  x == 5 or (y == 5 and y == 10)
#           = True  or (False and True)
#           = True  or False
#           = True
if x == 5 or y == 5 and y == 10:
    print("Yes")   # ← printed
else:
    print("No")

# To make it print "No", force the or-clause to be False:
if (x == 5 or y == 5) and y == 10:
    # (True or False) and True = True → still Yes
    print("Yes (grouped)")

# Guaranteed "No" — make x != 5 and override with parentheses:
x = 6
if (x == 5 or y == 5) and y == 10:
    print("Yes")
else:
    print("No")   # ← (False or False) and True = False
````

````python
num = int(input('Enter a number: '))

# lcm(6, 4) = 12; divisible by both ⟺ divisible by 12
if num % 6 == 0 and num % 4 == 0:
    print('Divisible by both 6 and 4')
elif num % 6 == 0:
    print('Divisible by 6 only')
elif num % 4 == 0:
    print('Divisible by 4 only')
else:
    print('Neither')

# Naive mistake: only checking `num % 12 == 0` for "both"
# fails for num = 24 → 24 % 12 == 0 ✓, 24 % 6 == 0 ✓, 24 % 4 == 0 ✓ → fine
# but fails conceptually if you write elif num % 12 == 0 as the ONLY "both" check.
````

### Q48 · String vs Integer Comparison
Why does `x == 10` fail when `x = "10"`? Why does this matter in data analysis?

````python
x = "10"   # string

if x == 10:
    print("Integer match")   # ← never reached
elif x == "10":
    print("String match")    # ← this runs
else:
    print("No match")

# In data analysis this matters because:
# - CSV files read all values as strings by default.
# - A column that looks like numbers may actually be strings.
# - Comparisons like `df['age'] > 18` will raise a TypeError or give wrong
#   results if the column was not converted to int/float first.
# Always check dtypes and cast appropriately: int(x) or float(x).
````

### Q49 · Boundary Edge Cases in Range Checks
Are these three forms equivalent?

````python
x = 10

# All three check the same range (exclusive boundaries: 1 < x < 10)
form1 = 1 < x < 10          # Python chained comparison
form2 = x > 1 and x < 10   # explicit logical and
form3 = 1 < x and x < 10   # verbose chained equivalent

print(form1, form2, form3)  # False False False for x=10 (10 is NOT < 10)

# ⚠️ Edge-case trap: all three EXCLUDE 1 and 10.
# If you want to include the boundaries, use: 1 <= x <= 10
print(1 <= x <= 10)  # True
````

### Q50 · Clamp a Value (if-elif + Ternary)
Clamp `num` to the range [0, 1000].

````python
num = 1200

# ── Readable if-elif-else ─────────────────────────────────────
if num < 0:
    clamped = 0
elif num > 1000:
    clamped = 1000
else:
    clamped = num
print(clamped)   # 1000

# ── Nested ternary (works, but harder to read) ────────────────
clamped = 0 if num < 0 else (1000 if num > 1000 else num)
print(clamped)   # 1000

# ── Pythonic built-in approach (for reference) ────────────────
clamped = max(0, min(1000, num))
print(clamped)   # 1000

# Conclusion: nested ternaries are syntactically valid but reduce readability.
# Prefer the if-elif or the max/min idiom in production code.
````

## Summary

| Concept | Key Takeaway |
|-------------------------------|----------------------------------------------------|
| `if` | Runs only when condition is `True` |
| `if-else` | Exactly one branch always executes |
| `if-elif-else` | First matching condition wins; rest are skipped |
| Nested `if` | Use sparingly; prefer `and`/`or` when possible |
| Ternary operator | One-liner for simple `if-else`; avoid nesting |
| Truthy / Falsy | `0`, `""`, `[]`, `{}`, `None` → `False`; rest → `True` |
| Short-circuit | `and` stops on first `False`; `or` stops on first `True` |
| Operator precedence | `not` > `and` > `or` |
| `=` vs `==` | `=` assigns; `==` compares |
