## 1. What is a Variable?

A **variable** is a name that stores a value in memory.

Think of it like a labelled box:

- The box has a name.
- You put a value inside it.
- You can use that name later to access the value.

```python
name = "Roshan"
age = 25
gender = "Male"

print(name)
print(age)
print(gender)
```

## 2. Variable Naming Rules

| Rule | Example |
|---|---|
| Must start with a letter or underscore `_` | `name`, `_name` |
| Can contain letters, numbers, and underscores | `name1`, `first_name` |
| Cannot start with a number | `1name` ❌ |
| Cannot use Python keywords | `if`, `for`, `while` ❌ |
| Variable names are case-sensitive | `age`, `Age`, and `AGE` are different |

```python
# Valid variable names
a1 = 5
_name1 = "Roshan"

print(a1)
print(_name1)

# Invalid variable name example:
# 1a = 5   # This will give a SyntaxError because variable names cannot start with a number.

age = 25
Age = 26
AGE = 27

# All three variables are different because Python is case-sensitive.
print(age)
print(Age)
print(AGE)
```

## 3. Data Types in Python

Every value in Python has a data type. Python automatically detects the type based on the value.

| Data Type | Meaning | Example |
|---|---|---|
| `int` | Whole number | `25` |
| `float` | Decimal number | `87.65` |
| `str` | Text/string | `"Roshan"` |
| `bool` | True/False value | `True` |

> Other useful types that come later: `complex` and `NoneType`.

```python
name = "Roshan"
age = 25
gender = "Male"
is_student = True
marks = 87.65

print(type(name))
print(type(age))
print(type(gender))
print(type(is_student))
print(type(marks))
```

## 4. The `print()` Function

The `print()` function is used to display output on the screen.

You can print:

- A single value
- Multiple values
- Expressions
- Data types

```python
name = "Roshan"
age = 25
gender = "Male"

# Printing a single value
print("Hello")

# Printing multiple values
print("Hello", name, "Your age is", age, "and you are", gender)

# Printing expressions
print(10 + 5)

# Printing data type
print(type(3.12))
```

## 5. `print()` with `sep`, `end`, and f-strings

By default:

- `print()` adds a space between multiple values.
- `print()` moves to the next line after printing.

We can customize this using:

- `sep` → changes the separator between values
- `end` → changes what appears at the end
- f-string → makes formatted output easier

```python
name = "Roshan"
age = 25
gender = "Male"

# Default printing
print(name)
print(age)
print(gender)

# sep changes the separator between values
print(name, age, gender, sep="_")
print(name, age, gender, sep="|")

# end changes what is printed at the end
print("Downloading...", end="")
print("Done")

# f-string formatting
print(f"Your name is {name}. You are {age} years old and you are {gender}.")
print(f"After 2 years, you will be {age + 2} years old.")
```

## 6. Type Conversion: Implicit vs Explicit

### Implicit Type Conversion

Python automatically converts one data type into another when it is safe to do so.

```python
x = 5       # int
y = 2.0     # float

z = x + y   # Python converts x into float automatically

print(z)
print(type(z))
```

### Explicit Type Conversion

Explicit conversion means we manually convert one data type into another.

Common functions:

- `int()`
- `float()`
- `str()`
- `bool()`

```python
num1 = "100"
num2 = "200"

num3 = int(num1)
num4 = int(num2)

print(num1 + num2)   # String concatenation
print(num3 + num4)   # Integer addition

# Invalid conversion example:
# int("100abc")  # This will give ValueError because the string is not a pure number.
```

## 8. Escape Sequences

Escape sequences start with a backslash `\`. They are used to represent special characters inside strings.

| Escape Sequence | Meaning |
|---|---|
| `\n` | New line |
| `\t` | Tab space |
| `\\` | Backslash |
| `\"` | Double quote |
| `\'` | Single quote |

```python
print("Hello\nWorld")
print("Hello\tWorld")
print("Hello\\World")
print("Hello \"World\"")
print("Hello \'World\'")
```

## 9. Arithmetic Operators

Arithmetic operators are used to perform mathematical operations.

| Operator | Meaning | Example |
|---|---|---|
| `+` | Addition | `10 + 20` |
| `-` | Subtraction | `20 - 10` |
| `*` | Multiplication | `20 * 10` |
| `/` | Division | `20 / 10` |
| `//` | Floor division | `20 // 3` |
| `%` | Modulus/remainder | `20 % 7` |
| `**` | Power/exponent | `2 ** 5` |

```python
print(10 + 20)
print(20 - 10)
print(20 * 10)
print(20 / 10)    # Division always returns a float
print(20 // 3)    # Floor division
print(20 % 7)     # Remainder
print(2 ** 5)     # Power
```

## 10. Operator Precedence

When multiple operators are used in one expression, Python follows a priority order.

Order from highest to lowest:

1. `**` → Exponentiation
2. `*`, `/`, `//`, `%` → Multiplication and division operators
3. `+`, `-` → Addition and subtraction

Use parentheses `()` when you want to control the order clearly.

```python
print(2 + 3 * 4)        # 14, because multiplication happens first
print(10 - 2 ** 3)      # 2, because exponent happens first
print(10 // 2 + 3)      # 8

# Parentheses can change the order
print((2 + 3) * 4)      # 20
print((10 - 2) ** 3)    # 512
```

## 11. Comparison Operators

Comparison operators compare two values and return either `True` or `False`.

| Operator | Meaning |
|---|---|
| `==` | Equal to |
| `!=` | Not equal to |
| `>` | Greater than |
| `<` | Less than |
| `>=` | Greater than or equal to |
| `<=` | Less than or equal to |

```python
a = 10
b = 10

print(a == b)
print(a > b)
print(a < b)
print(a <= b)
print(a >= b)
print(a != b)
```

## 12. Logical Operators

Logical operators are used to combine multiple conditions.

| Operator | Meaning |
|---|---|
| `and` | Returns `True` only when all conditions are true |
| `or` | Returns `True` when at least one condition is true |
| `not` | Reverses the result |

```python
print(10 > 5 and 40 == 40 and 50 < 100)    # True
print(10 > 5 and 40 == 40 and 50 == 100)   # False

print(10 < 5 or 40 < 40 or 50 < 100)       # True

print(not 100 > 5)                         # False
```

## 13. Assignment Operators

Assignment operators are shortcuts for updating a variable value.

| Operator | Meaning | Same As |
|---|---|---|
| `=` | Assign value | `x = 10` |
| `+=` | Add and assign | `x = x + 5` |
| `-=` | Subtract and assign | `x = x - 5` |
| `*=` | Multiply and assign | `x = x * 5` |
| `/=` | Divide and assign | `x = x / 5` |
| `//=` | Floor divide and assign | `x = x // 5` |
| `%=` | Modulus and assign | `x = x % 5` |
| `**=` | Power and assign | `x = x ** 5` |

```python
x = 10

x += 5
print(x)

x -= 3
print(x)

x *= 2
print(x)

x /= 4
print(x)
```
